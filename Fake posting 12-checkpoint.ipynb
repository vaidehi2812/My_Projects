```python
def department_info(department_name):
    text_IT = data[data['department']==department_name]["preprocessed"].str.lower().str.replace(r'\|', ' ').str.cat(sep=' ')
    tokens = nltk.tokenize.word_tokenize(text_IT)
    word_dist = nltk.FreqDist(tokens)
    stop_words = stopwords.words('english')
    punctuations = ['(',')',';',':','[',']',',','!','?']
    keywords = [word for word in tokens if not word in stop_words and not word in punctuations]
    #print(keywords)
    counts = Counter(keywords)
    #print("")
    print(department_name, counts)
```

```python
def salary_dep(department_name):
    data_min= data[data['department']==department_name]['min_salary']
    #print(data_min)
    print("Department name: ",department_name,"\n")
    data_min =  np.array([data_min])
    data_min = np.average(data_min)
    print("Salary Min: ",data_min)
    
    data_max= data[data['department']==department_name]['max_salary']
    #print(data_max)
    data_max =  np.array([data_max])
    data_max = np.average(data_max)
    
    print("Salary Max: ",data_max)
    
    print("Total salary of ",department_name,"is:",(data_max+data_min)/2)
    
    data.loc[data['department'].str.contains(department_name),'salary'] = (data_max+data_min)/2
```
